```python
# 1: by mapping chars to words in which they occur

from collections import defaultdict


class Solution(object):
    def maxProduct(self, words):
        """
        :type words: List[str]
        :rtype: int
        """

        # remember in what word each char occures

        have_chars = defaultdict(set)

        for wi, w in enumerate(words):
            for c in w:
                have_chars[c].add(wi)

        # compare pairwise

        mx = 0

        for wi1, w1 in enumerate(words):

            # which words to skip
            skip = set()
            for c in w1:
                for i in have_chars[c]:
                    skip.add(i)

            for wi2, w2 in enumerate(words):
                if wi1 != wi2 and not wi2 in skip:
                    mx = max(len(w1) * len(w2), mx)

        return mx
```

```python
# 2: by creating but masks for each words with bits corresponding to chars occuring


class Solution(object):
    def maxProduct(self, words):
        """
        :type words: List[str]
        :rtype: int
        """

        # convert words to ins with 1s set at i=0..26 if char i occurs

        word_masks = []  # word bitmasks
        max_l_for_mask = {}  # max word length for each mask

        for w in words:
            mask = 0

            for c in w:
                mask |= 1 << ord(c) - 97

            if max_l_for_mask.get(mask, 0) < len(w):
                max_l_for_mask[mask] = len(w)

        mx = 0
        masks = list(max_l_for_mask.keys())

        for i1 in range(len(masks)):
            for i2 in range(i1 + 1, len(masks)):
                m1, m2 = masks[i1], masks[i2]
                if m1 & m2 == 0:  # no same bits are set
                    mx = max(mx, max_l_for_mask[m1] * max_l_for_mask[m2])

        return mx
```
