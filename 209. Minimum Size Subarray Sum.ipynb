```python
class Solution(object):
    def minSubArrayLen(self, s, nums):
        """
        :type s: int
        :type nums: List[int]
        :rtype: int
        """

        a = 0
        cur_sum = 0
        r = float('inf')

        for b in range(len(nums)): # move b
            cur_sum += nums[b]

            while cur_sum >= s: # move a
                r = min(r, b - a + 1)
                cur_sum -= nums[a]
                a += 1

        return r if r < float('inf') else 0
```
