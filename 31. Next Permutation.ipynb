```python
# https://leetcode.com/articles/next-permutation/


class Solution(object):
    def nextPermutation(self, n):

        # find rightmost element where n[i-1] < n[i]
        for i in reversed(range(len(n))):
            if i > 0 and n[i] > n[i - 1]:
                is_sorted = False
                break

        # if leftmost element is the greatest, then
        # the sequence is sorted in descending order
        if 0 == i:
            n.reverse()
            return

        # swap n[i-1] with rightmost element which is bigger
        for j in reversed(range(i, len(n))):
            if n[j] > n[i - 1]:
                n[j], n[i - 1] = n[i - 1], n[j]
                break

        # now to get the smalest permutation, numbers
        # right of n[i-1] need to be sorted in asc order
        # (they're in desc order currently)

        a = i
        b = len(n) - 1

        while a < b:
            n[a], n[b] = n[b], n[a]
            a += 1
            b -= 1
```
