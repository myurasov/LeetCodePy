```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# The guess API is already defined for you.
# @param num, your guess
# @return -1 if my number is lower, 1 if my number is higher, otherwise return 0
# def guess(num):

NUM = 0

def guess(num):
    if num > NUM: return -1
    if num < NUM: return 1
    return 0
```

```python
cc = []

class Solution(object):
    def guessNumber(self, n):
        """
        :type n: int
        :rtype: int
        """
        
        mi = 0
        ma = n

        c = int(ma / 2)
        g = -1

        while g != 0:
        
            g = guess(c)
            
            cc.append(c)
            
            if g > 0:
                mi = c
            elif g < 0:
                ma = c
            
            d = max(1, int((ma - mi) / 2)) # step is a half of diff between max and min, but no less than 0
            d *= g # direction of the step
            c += d
                
                
        return c
```

```python
NUM = 11111
s = Solution()
assert(NUM == s.guessNumber(100000))

plt.plot(cc)
```

```python
cc = []
NUM = 1
s = Solution()
assert(NUM == s.guessNumber(1000))

plt.plot(cc)
```
