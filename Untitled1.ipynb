```python
from collections import OrderedDict


class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """

        self.res = 0
        self.endWord = endWord
        self.wordList = set(wordList)
        self.dfs(beginWord)
        return self.res

    def dfs(self, root, d=0, path=set()):

        if root not in path:

            if root == self.endWord:
                if self.res == 0: self.res = d + 1
                elif self.res > d + 1: self.res = d + 1

            children = self.getChildren(root, self.wordList, path)

            for c in children:
                new_path = path.copy()
                new_path.add(root)
                self.dfs(c, d + 1, new_path)

    def getChildren(self, word, candidates, used):

        res = []

        for cw in candidates:
            if cw in used: continue

            diff = 0

            for i in range(len(word)):
                if word[i] != cw[i]: diff += 1
                if diff > 1: break

            if diff == 1: res.append(cw)

        return res
```

```python
from collections import *
from heapq import *


class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """

        nodes = self.createGraph(beginWord, wordList)
#         d = self.findShortestDistance(beginWord, endWord, nodes)
#         return d + 1 if d < float('inf') else 0
        self.n = nodes
        self.b = beginWord
        self.e = endWord
        
    def a(self):
        d = self.findShortestDistance(self.b, self.e, self.n)
        return d + 1 if d < float('inf') else 0

    # find shortest dist with Dijkstra algorithm
    # https://en.wikipedia.org/wiki/Dijkstra%27s_algorithm
    def findShortestDistance(self, beginWord, endWord, nodes):

        current = nodes[beginWord]
        current['distance'] = 0
        
        # todo: unvisited as minheap (dist, key)

        while True:

            if current['key'] != endWord:

                # calc tentative distances to neighbors
                for w in current['neighbors']:
                    neighbor = nodes[w]
                    neighbor['distance'] = min(neighbor['distance'],
                                               current['distance'] + 1)

                current['visited'] = True

                # find next unvisited node with min distance

                min_distance = float('inf')

                for key in nodes:
                    node = nodes[key]
                    if not node['visited'] and node['distance'] < min_distance:
                        current = node
                        min_distance = current['distance']

                if min_distance == float('inf') or current['key'] == endWord:
                    return min_distance

    def createGraph(self, beginWord, wordList):

        nodes = {
            x: {
                'key': x,
                'neighbors': [],
                'distance': float('inf'),
                'visited': False
            }
            for x in [beginWord] + wordList
        }

        # connect nodes
        for w1 in nodes:
            for w2 in nodes:

                # find diff between w1 and w2
                diff = 0
                for i in range(len(w1)):
                    if w1[i] != w2[i]:
                        diff += 1
                    if diff > 1: break

                if diff == 1:
                    nodes[w1]['neighbors'].append(w2)

        return nodes
```
